```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading the csv file and storing it in a dataframe called df
df = pd.read_csv('C:/Users/Kumar/OneDrive/Documents/Machine_Learning_Shahram/Classification Problem/gender_classification_v7.csv')
df
```

```python
df.columns
```

```python
X = df[['long_hair', 'forehead_width_cm', 'nose_wide',
       'nose_long', 'lips_thin', 'distance_nose_to_lip_long', 'gender']]
    
y = df['forehead_height_cm']
```

```python
n_cols = X.select_dtypes(include=np.number).columns.to_list()
c_cols = X.select_dtypes('object').columns.to_list()
```

```python
n_cols, c_cols
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[n_cols])
df[n_cols]=scaler.transform(df[n_cols])
```

```python
from sklearn.preprocessing import LabelEncoder
df[c_cols] = LabelEncoder().fit_transform(df[c_cols])
df[c_cols]
```

```python
X = df[n_cols + c_cols]
y = df['forehead_height_cm']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
#ML Models
from sklearn.linear_model import LinearRegression

from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
#Tree
from sklearn.tree import DecisionTreeRegressor
# Ensembling Techniques
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor

#Scores & Metrics
from sklearn.model_selection import cross_val_score
```

```python
X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=30)
```

```python
model_reg = LinearRegression()
model_reg.fit(X_train, y_train)
#X_test = selector.transform(X_test)
predict = model_reg.predict(X_test)
print('Intercept is :',model_reg.intercept_)
print('Coefficient is :' ,model_reg.coef_)
print('Linear Regression Model Score -',model_reg.score(X_test,y_test))
```

```python
np.mean((predict-y_test)**2)
```

```python
pd.DataFrame({'actual':y_test,
            'prediction':predict,
            'diff':(y_test-predict)})
```

```python
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train, y_train)
predict = dt_reg.predict(X_test)
print('Decision Tree Regressor Model Score -',dt_reg.score(X_test,y_test))
```

```python
sv_reg = SVR()
sv_reg.fit(X_train, y_train)
predict = dt_reg.predict(X_test)
print('Support Vector Regressor Model Score -',sv_reg.score(X_test,y_test))
```

```python
ada_boost = AdaBoostRegressor(base_estimator=LinearRegression(), n_estimators=450)
ada_boost.fit(X_train, y_train)
predict = ada_boost.predict(X_test)
print('AdaBoost Regressor Model Score -',ada_boost.score(X_test,y_test))
```
